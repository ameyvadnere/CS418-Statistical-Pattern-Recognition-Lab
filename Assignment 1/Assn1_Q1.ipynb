```python
# We import the necessary libraries.

import numpy as np    # For numerical computations
import seaborn as sns # For plotting
import matplotlib.pyplot as plt
import matplotlib
```

```python
# Function for plotting sample points generated from a distribution.

def plot(x):
    sns.set_style('darkgrid')
    fig = sns.histplot(x=x)
    fig.set(xlabel='x (Random Variable)', title='Histogram of x')
    plt.show()
```

```python
# We generate samples from uniform distribution between 0 and 1.

u_100 = np.random.uniform(0, 1, 100)     # 100 uniform points
u_1000 = np.random.uniform(0, 1, 1000)   # 1000 uniform points
u_10000 = np.random.uniform(0, 1, 10000) # 10000 uniform points

plot(u_100)
plot(u_1000)
plot(u_10000)
```

```python
# We generate samples from Gaussian distribution between 0 and 1 with mean=0, std=1

g_100 = np.random.normal(0, 1, 100)     # 100 points
g_1000 = np.random.normal(0, 1, 1000)   # 1000 points
g_10000 = np.random.normal(0, 1, 10000) # 10000 points

plot(g_100)
plot(g_1000)
plot(g_10000)
```

```python
# We now generate samples from Exponential distribution with rate=1

e_100 = np.random.exponential(1, 100)    # 100 points
e_1000 = np.random.exponential(1, 1000)  # 1000 points
e_10000 = np.random.exponential(1, 10000)# 10000 points

plot(e_100)
plot(e_1000)
plot(e_10000)

```
